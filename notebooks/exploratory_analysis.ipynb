## Colunas que permaneceram

DT_NOTIFIC = data de notificação, importante para o problema

AVE_SUINO = essa coluna se refere ao contato de humanos com aves ou suínos, é interessante para o problema porque surtos síndromes respiratórias já ocorreram no passado por causa desse contato, como, por exemplo, o vírus H1N1

FEBRE = febre, esse coluna é interessante para o problema

TOSSE = tosse, não é interessante, porque a tosse não é exclusiva de síndromes respiratórias  

GARGANTA = dor de garganta, não é interssante 

DISPNEIA = falta de ar, é interessante para o problema pois é um fator crítico para identificar uma síndrome resíratória aguda grave  

FATOR_RISC = fator de risco, é interssante para a análise

ANTIVIRAL = antiviral, é interassante para o problema. Principalmente em um momento de surtos ou pandemias para identificar se o antiviral está surtindo efeito na evolução do paciente 

UTI = uti, é interessante para o problema pois indica a gravidade da doença 

EVOLUCAO = evolução do quadro do paciente, é interessante para o problema pois indica a gravidade da doença 

VACINA = vacina, é interessante para o problema 

SURTO_SG = surto de síndrome respiratória, é interessante par o problema pois é uma indicação de ploriferação de doenças
